#1) Data Cleaning

Find all the null values in the dataset.
If there is any null value in the column(integer), fill it with the mean of that column

```
car.isnull().sum()
```

```
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace=True)



car['EngineSize'].fillna(car['EngineSize'].mean(), inplace=True)
car['Horsepower'].fillna(car['Horsepower'].mean(), inplace=True)
car['MPG_City'].fillna(car['MPG_City'].mean(), inplace=True)
car['MPG_Highway'].fillna(car['MPG_Highway'].mean(), inplace=True)
car['Weight'].fillna(car['Weight'].mean(), inplace=True)
car['Wheelbase'].fillna(car['Wheelbase'].mean(), inplace=True)
car['Length'].fillna(car['Length'].mean(), inplace=True)
```

```
car.isnull().sum()
```

```
car['Cylinders'].isnull().sum()
```

```
car.fillna('Empty', inplace=True)
```

#3) Filtering
Show all the records where origin is Asia or Europe

```
car.head(2)
```

```
#using basic filtering method
car[(car['Origin'] == 'Asia') | (car['Origin'] == 'Europe')]
```

```
# using is in function
car[car['Origin'].isin(['Asia', 'Europe'])]
```

#4) Remove all the unwanted records

Remove all the records(rows) where weight is above 4000

```
car.head(2)
```

```
car[~(car['Weight'] > 4000)]
```

#5) Applying a function on a column

Increase all the values of 'MPG_City' column by 3

```
car['MPG_City'] = car['MPG_City'].apply(lambda x: x+3)
```
